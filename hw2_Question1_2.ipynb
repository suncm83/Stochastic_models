```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Initialize
c1=1
c2=0.005
c3=0.6
v1=[1,0]
v2=[-1,1]
v3=[0,-1]
```

```python
# Assume Y1 and Y2 initial sizes are 10
Y1=[10]
Y2=[10]

t=[0]
t_end=10
i=0

#Gillespe Algorithm
while (t[i]<=t_end) & (Y1[i]>0) & (Y2[i]>0):
#Evaluate a1, a2, a3 and asum
    a1=c1*Y1[i]
    a2=c2*Y1[i]*Y2[i]
    a3=c3*Y2[i]
    asum=a1+a2+a3
#Two independent uniform (0,1) random numbers
    e1 = np.random.uniform(0,1,1)
    e2 = np.random.uniform(0,1,1)

#find the smallest j     
    aj=a1
    j=1
    if aj<=(e1*asum):
        aj=aj+a2
        j=2
        if aj<=(e1*asum):
            aj=a1+a2+a3
            j=3
#Set tau
    tau=(np.log(1/e2))/asum
    
#Update t and Y1, Y2    
    t.append(t[i]+tau)
    
    if j==1:
        Y1.append(Y1[i]+v1[0])
        Y2.append(Y2[i]+v1[1])
    if j==2:
        Y1.append(Y1[i]+v2[0])
        Y2.append(Y2[i]+v2[1]) 
    if j==3:
        Y1.append(Y1[i]+v3[0])
        Y2.append(Y2[i]+v3[1])        
    i+=1
    
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle(f"c1={c1},c2={c2},c3={c3}")
plt.xlabel("Time")
plt.ylabel("Concentration")
ax1.plot(t,Y1,'-r')
ax1.legend(["Y1"])
ax2.plot(t,Y2,'-b')
ax2.legend(["Y2"])

        
```
