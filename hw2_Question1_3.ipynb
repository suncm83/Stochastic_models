```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Initialize
c1=np.arange(0.2,2,0.2)
c2=np.arange(0.005,0.045,0.005)
c3=np.arange(0.2,2,0.2)
v1=[1,0]
v2=[-1,1]
v3=[0,-1]

```

```python
#Increase c1 gradually
Y1={}
Y2={}
t={}

for k in range(len(c1)):

# Assume Y1 and Y2 initial sizes are 10
    Y1["c1_"+str(c1[k])]=[10]
    Y2["c1_"+str(c1[k])]=[10]
    t["c1_"+str(c1[k])]=[0]

    t_end=10
    i=0

#Gillespe Algorithm
    while (t["c1_"+str(c1[k])][i]<=t_end) & (Y1["c1_"+str(c1[k])][i]>0) & (Y2["c1_"+str(c1[k])][i]>0):
#Evaluate a1, a2, a3 and asum
        a1=c1[k]*Y1["c1_"+str(c1[k])][i]
        a2=0.005*Y1["c1_"+str(c1[k])][i]*Y2["c1_"+str(c1[k])][i]
        a3=0.6*Y2["c1_"+str(c1[k])][i]
        asum=a1+a2+a3
#Two independent uniform (0,1) random numbers
        e1 = np.random.uniform(0,1,1)
        e2 = np.random.uniform(0,1,1)

#find the smallest j     
        aj=a1
        j=1
        if aj<=(e1*asum):
            aj=aj+a2
            j=2
            if aj<=(e1*asum):
                aj=a1+a2+a3
                j=3
#Set tau
        tau=(np.log(1/e2))/asum
    
#Update t and Y1, Y2    
        t["c1_"+str(c1[k])].append(t["c1_"+str(c1[k])][i]+tau)
    
        if j==1:
            Y1["c1_"+str(c1[k])].append(Y1["c1_"+str(c1[k])][i]+v1[0])
            Y2["c1_"+str(c1[k])].append(Y2["c1_"+str(c1[k])][i]+v1[1])
        if j==2:
            Y1["c1_"+str(c1[k])].append(Y1["c1_"+str(c1[k])][i]+v2[0])
            Y2["c1_"+str(c1[k])].append(Y2["c1_"+str(c1[k])][i]+v2[1]) 
        if j==3:
            Y1["c1_"+str(c1[k])].append(Y1["c1_"+str(c1[k])][i]+v3[0])
            Y2["c1_"+str(c1[k])].append(Y2["c1_"+str(c1[k])][i]+v3[1])        
        i+=1
    
fig, axs = plt.subplots(len(c1), 2,figsize=(15,35))
plt.xlabel("Time")
plt.ylabel("Concentration")
for k in range(len(c1)):
    axs[k,0].title.set_text(f'c1={c1[k]}')
    axs[k,1].title.set_text(f'c1={c1[k]}')
    axs[k,0].plot(t["c1_"+str(c1[k])],Y1["c1_"+str(c1[k])],'-r')
    axs[k,0].legend(["Y1"])
    axs[k,1].plot(t["c1_"+str(c1[k])],Y2["c1_"+str(c1[k])],'-r')
    axs[k,1].legend(["Y2"])
```
