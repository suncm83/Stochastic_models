```python
import numpy as np
import matplotlib.pyplot as plt
import operator
```

```python
A={}
B={}
C={}
X={}
Y={}
t={}


state_A_sorted={}
state_B_sorted={}
state_C_sorted={}
state_X_sorted={}
state_Y_sorted={}


v1=[-1,0,0,1,0]
v2=[0,0,0,-1,0]
v3=[0,-1,0,0,1]
v4=[0,0,0,0,-1]
v5=[0,0,1,-1,-1]

k=[10,1000]
alpha1=[10**(-6),10**(-4)]
alpha2=[10**(-5),10**(-3)]
ka=[10**(-5),10**(-3)]

state_A_sorted_range={}
state_B_sorted_range={}
state_C_sorted_range={}
state_X_sorted_range={}
state_Y_sorted_range={}

state_A_sorted_prob={}
state_B_sorted_prob={}
state_C_sorted_prob={}
state_X_sorted_prob={}
state_Y_sorted_prob={}

T={}

for n in range(1,3,1):
    
    state_A={}
    state_B={}
    state_C={}
    state_X={}
    state_Y={}
    
    state_A_sorted_range["Case"+str(n)]=[]
    state_B_sorted_range["Case"+str(n)]=[]
    state_C_sorted_range["Case"+str(n)]=[]
    state_X_sorted_range["Case"+str(n)]=[]
    state_Y_sorted_range["Case"+str(n)]=[]
    
    state_A_sorted_prob["Case"+str(n)]=[]
    state_B_sorted_prob["Case"+str(n)]=[]
    state_C_sorted_prob["Case"+str(n)]=[]
    state_X_sorted_prob["Case"+str(n)]=[]
    state_Y_sorted_prob["Case"+str(n)]=[]
    
    
    T["Case"+str(n)]=0.0
    
    for nn in range (1000):
    
    
        A["Case"+str(n)]=[50]
        B["Case"+str(n)]=[100]
        C["Case"+str(n)]=[200]
        X["Case"+str(n)]=[5000]
        Y["Case"+str(n)]=[5000]
        t["Case"+str(n)]=[0]
        t_end=5
        i=0



        while (t["Case"+str(n)][i]<=t_end) & (A["Case"+str(n)][i]>0) & (B["Case"+str(n)][i]>0) & (C["Case"+str(n)][i]>0) & (X["Case"+str(n)][i]>0) & (Y["Case"+str(n)][i]>0):
            a1=k[n-1]*A["Case"+str(n)][i]
            a2=alpha1[n-1]*X["Case"+str(n)][i]

            a3=k[n-1]*B["Case"+str(n)][i]
            a4=alpha2[n-1]*Y["Case"+str(n)][i]
            a5=ka[n-1]*X["Case"+str(n)][i]*Y["Case"+str(n)][i]
            asum=a1+a2+a3+a4+a5
            e1 = np.random.uniform(0,1,1)
            e2 = np.random.uniform(0,1,1)
            aj=a1
            j=1
            if aj<=(e1*asum):
                aj=aj+a2
                j=2
                if aj<=(e1*asum):
                    aj=aj+a3
                    j=3
                    if aj<=(e1*asum):
                        aj=aj+a4
                        j=4
                        if aj<=(e1*asum):
                            aj=aj+a5
                            j=5
            tau=(np.log(1/e2))/asum
            t["Case"+str(n)].append(t["Case"+str(n)][i]+tau)
            tempA=str(round(A["Case"+str(n)][i]))
            tempB=str(round(B["Case"+str(n)][i]))
            tempC=str(round(C["Case"+str(n)][i]))
            tempX=str(round(X["Case"+str(n)][i]))
            tempY=str(round(Y["Case"+str(n)][i]))

            if tempA in state_A:
                state_A[tempA]=[round(A["Case"+str(n)][i]),state_A[tempA][1]+tau]
            else:
                state_A[tempA]=[round(A["Case"+str(n)][i]),tau]

            if tempB in state_B:
                state_B[tempB]=[round(B["Case"+str(n)][i]),state_B[tempB][1]+tau]
            else:
                state_B[tempB]=[round(B["Case"+str(n)][i]),tau]

            if tempC in state_C:
                state_C[tempC]=[round(C["Case"+str(n)][i]),state_C[tempC][1]+tau]
            else:
                state_C[tempC]=[round(C["Case"+str(n)][i]),tau]

            if tempX in state_X:
                state_X[tempX]=[round(X["Case"+str(n)][i]),state_X[tempX][1]+tau]
            else:
                state_X[tempX]=[round(X["Case"+str(n)][i]),tau]                

                
            if tempY in state_Y:
                state_Y[tempY]=[round(Y["Case"+str(n)][i]),state_Y[tempY][1]+tau]
            else:
                state_Y[tempY]=[round(Y["Case"+str(n)][i]),tau]                

            T["Case"+str(n)]=T["Case"+str(n)]+tau
            if j==1:
                A["Case"+str(n)].append(A["Case"+str(n)][i]+v1[0])
                B["Case"+str(n)].append(B["Case"+str(n)][i]+v1[1])
                C["Case"+str(n)].append(C["Case"+str(n)][i]+v1[2])
                X["Case"+str(n)].append(X["Case"+str(n)][i]+v1[3])
                Y["Case"+str(n)].append(Y["Case"+str(n)][i]+v1[4])
            if j==2:
                A["Case"+str(n)].append(A["Case"+str(n)][i]+v2[0])
                B["Case"+str(n)].append(B["Case"+str(n)][i]+v2[1])
                C["Case"+str(n)].append(C["Case"+str(n)][i]+v2[2])
                X["Case"+str(n)].append(X["Case"+str(n)][i]+v2[3])
                Y["Case"+str(n)].append(Y["Case"+str(n)][i]+v2[4])
            if j==3:
                A["Case"+str(n)].append(A["Case"+str(n)][i]+v3[0])
                B["Case"+str(n)].append(B["Case"+str(n)][i]+v3[1])
                C["Case"+str(n)].append(C["Case"+str(n)][i]+v3[2])
                X["Case"+str(n)].append(X["Case"+str(n)][i]+v3[3])
                Y["Case"+str(n)].append(Y["Case"+str(n)][i]+v3[4]) 
            if j==4:
                A["Case"+str(n)].append(A["Case"+str(n)][i]+v4[0])
                B["Case"+str(n)].append(B["Case"+str(n)][i]+v4[1])
                C["Case"+str(n)].append(C["Case"+str(n)][i]+v4[2])
                X["Case"+str(n)].append(X["Case"+str(n)][i]+v4[3])
                Y["Case"+str(n)].append(Y["Case"+str(n)][i]+v4[4])
            if j==5:
                A["Case"+str(n)].append(A["Case"+str(n)][i]+v5[0])
                B["Case"+str(n)].append(B["Case"+str(n)][i]+v5[1])
                C["Case"+str(n)].append(C["Case"+str(n)][i]+v5[2])
                X["Case"+str(n)].append(X["Case"+str(n)][i]+v5[3])
                Y["Case"+str(n)].append(Y["Case"+str(n)][i]+v5[4])        

            i+=1
        
    state_A_sorted=sorted(state_A.items(), key=operator.itemgetter(1,0),reverse=False)
    state_B_sorted=sorted(state_B.items(), key=operator.itemgetter(1,0),reverse=False)
    state_C_sorted=sorted(state_C.items(), key=operator.itemgetter(1,0),reverse=False)
    state_X_sorted=sorted(state_X.items(), key=operator.itemgetter(1,0),reverse=False)
    state_Y_sorted=sorted(state_Y.items(), key=operator.itemgetter(1,0),reverse=False)    

    for ii in range(len(state_A_sorted)):
        state_A_sorted_range["Case"+str(n)].append(state_A_sorted[ii][1][0])
        state_A_sorted_prob["Case"+str(n)].append(float(state_A_sorted[ii][1][1]/T["Case"+str(n)]))

    for ii in range(len(state_B_sorted)):
        state_B_sorted_range["Case"+str(n)].append(state_B_sorted[ii][1][0])
        state_B_sorted_prob["Case"+str(n)].append(float(state_B_sorted[ii][1][1]/T["Case"+str(n)]))

    for ii in range(len(state_C_sorted)):
        state_C_sorted_range["Case"+str(n)].append(state_C_sorted[ii][1][0])
        state_C_sorted_prob["Case"+str(n)].append(float(state_C_sorted[ii][1][1]/T["Case"+str(n)]))

    for ii in range(len(state_X_sorted)):
        state_X_sorted_range["Case"+str(n)].append(state_X_sorted[ii][1][0])
        state_X_sorted_prob["Case"+str(n)].append(float(state_X_sorted[ii][1][1]/T["Case"+str(n)]))

    for ii in range(len(state_Y_sorted)):
        state_Y_sorted_range["Case"+str(n)].append(state_Y_sorted[ii][1][0])
        state_Y_sorted_prob["Case"+str(n)].append(float(state_Y_sorted[ii][1][1]/T["Case"+str(n)]))


fig, (ax1,ax2) = plt.subplots(1, 2,figsize=(10,10))
plt.xlabel("Concentration")
plt.ylabel("probability")
ax1.set_title(f'Case1 - A={A["Case1"][0]};B={B["Case1"][0]},C={C["Case1"][0]},X={X["Case1"][0]},Y={Y["Case1"][0]}')
ax1.plot(state_X_sorted_range["Case1"],state_X_sorted_prob["Case1"],'-r')
ax1.legend(["X"])
ax2.set_title(f'Case2 - A={A["Case2"][0]};B={B["Case2"][0]},C={C["Case2"][0]},X={X["Case2"][0]},Y={Y["Case2"][0]}')
ax2.plot(state_X_sorted_range["Case2"],state_X_sorted_prob["Case2"],'-r')
ax2.legend(["X"]) 

        

# fig, (ax1,ax2) = plt.subplots(1, 2,figsize=(10,10))
# plt.xlabel("Time")
# plt.ylabel("Concentration")
# ax1.set_title(f'Case1 - A={A["Case1"][0]};B={B["Case1"][0]},C={C["Case1"][0]},X={X["Case1"][0]},Y={Y["Case1"][0]}')
# ax1.plot(t["Case1"],X["Case1"],'-r')
# ax1.plot(t["Case1"],Y["Case1"],'--b')
# ax1.legend(["X","Y"])
# ax2.set_title(f'Case2 - A={A["Case2"][0]};B={B["Case2"][0]},C={C["Case2"][0]},X={X["Case2"][0]},Y={Y["Case2"][0]}')
# ax2.plot(t["Case2"],X["Case2"],'-r')
# ax2.plot(t["Case2"],Y["Case2"],'--b')
# ax2.legend(["X","Y"])        
        
```
